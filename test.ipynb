```python
import torch
from Generator import Generator
generator = Generator(100, 64, 3, 0)
generator.load_state_dict(torch.load('./models/generator.pth', map_location=torch.device("cpu")))


input = torch.randn(64, 100, 1, 1)
out = generator(input)
out.shape
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(nrows=8, ncols=8, figsize=(8, 8),
                        subplot_kw={'xticks': [], 'yticks': []})

for i, ax in enumerate(axs.flat):
    ax.imshow((out.detach()[i].permute(1, 2, 0)))
plt.tight_layout()
plt.show()
```
